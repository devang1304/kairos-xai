Combine the dataset for the date 4/3 from the graph_4_3 folder.

```python
import os
import pandas as pd

# Define the folder path
folder_path = '/Users/hetavpatel/Desktop/Data Science/Grad DS Work/DSCI 789 Explainable AI/XAI-Cybersec/XAI-CyberSec/tgnnexplainer/kairosSample/graph_4_3'

# Initialize an empty list to store data
combined_data = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            for line in file:
                combined_data.append(eval(line.strip()))

# Write the combined data to a giant text file
giant_text_file_path = os.path.join(folder_path, 'combined_data.txt')
with open(giant_text_file_path, 'w') as giant_file:
    for item in combined_data:
        giant_file.write(f"{item}\n")

# Convert the combined data to a pandas dataframe
df_combined = pd.DataFrame(combined_data)

# Save the dataframe to a CSV file
csv_file_path = 'combined_data_4_3_Kairos.csv'
df_combined.to_csv(csv_file_path, index=False)

print(f"Combined data saved to {csv_file_path}")
```
